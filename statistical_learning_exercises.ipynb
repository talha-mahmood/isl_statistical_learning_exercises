# Exercise 9
This exercise involves the Auto data set studied in the lab. Make sure
 that the missing values have been removed from the data

```python
import numpy as np
# Load the auto.csv data
auto = pd.read_csv('Auto.csv')
auto.head()

```

 (a) Which of the predictors are quantitative, and which are qualitative?


Ans: Name, origin cylinders and year is qualitative and mpg, displacement, horsepower, weight, acceleration are quantitative




```python
auto.dtypes
```

```python
# Convert 'horsepower' to numeric
auto['horsepower'] = pd.to_numeric(auto['horsepower'], errors='coerce')
# Remove missing values
auto_clean = auto.dropna()
# Check for the number of null values in each column
null_values = auto_clean.isnull().sum()

# Display the count of null values
print(null_values)
```

(b) What is the range of each quantitative predictor? You can answer this using the min() and max() methods in numpy.

```python
# Define quantitative columns
quantitative_columns = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']

# Calculate the range (min and max)
range_min = auto_clean[quantitative_columns].min()
range_max = auto_clean[quantitative_columns].max()

# Display the range
range_values = pd.DataFrame({'Min': range_min, 'Max': range_max})
print(range_values)
```

 (c) What is the mean and standard deviation of each quantitative
 predictor?

```python
# Calculate the mean and standard deviation for quantitative predictors
mean_values = auto_clean[quantitative_columns].mean()
std_values = auto_clean[quantitative_columns].std()

# Display the mean and standard deviation
mean_std_values = pd.DataFrame({'Mean': mean_values, 'Std Dev': std_values})
print(mean_std_values)
```

(d) Now remove the 10th through 85th observations. What is the
 range, mean, and standard deviation of each predictor in the
 subset of the data that remains?


```python
# Reset the index
auto_clean.reset_index(drop=True, inplace=True)
# drop 10th through 85th observations
auto_clean_updated= auto_clean.drop(index=range(9,85))
# Calculate the mean and standard deviation for quantitative predictors
mean_values_updated = auto_clean_updated[quantitative_columns].mean()
std_values_updated = auto_clean_updated[quantitative_columns].std()

# Calculate the range (min and max)
range_min_updated = auto_clean_updated[quantitative_columns].min()
range_max_updated = auto_clean_updated[quantitative_columns].max()

# Display the mean and standard deviation
mean_std_range_values = pd.DataFrame({'Mean': mean_values_updated, 'Std Dev': std_values_updated,'Min': range_min_updated, 'Max': range_max_updated})
print(mean_std_range_values)

```

(e) Using the full data set, investigate the predictors graphically,
 using scatterplots or other tools of your choice. Create some plots
 highlighting the relationships among the predictors. Comment
 on your findings.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix to examine relationships between all variables
correlation_matrix = auto_clean[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration', 'year','origin','cylinders']].corr()

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatterplot of mpg vs horsepower to visualize relationship
sns.scatterplot(x='horsepower', y='mpg', data=auto_clean)
plt.title('MPG vs Horsepower')
plt.show()

# Scatterplot of mpg vs weight to visualize relationship
sns.scatterplot(x='weight', y='mpg', data=auto_clean)
plt.title('MPG vs Weight')
plt.show()

# Scatterplot of mpg vs displacement to visualize relationship
sns.scatterplot(x='displacement', y='mpg', data=auto_clean)
plt.title('MPG vs Displacement')
plt.show()

# Scatterplot of mpg vs cylinders to visualize relationship
sns.scatterplot(x='cylinders', y='mpg', data=auto_clean)
plt.title('MPG vs Cylinders')
plt.show()
```

Ans: The plots and the correlation matrix can help identify which predictors are useful for predicting mpg.

Example Interpretations:

*   MPG vs Horsepower: scatterplot shows a strong negative trend, it suggests that as horsepower increases, mpg decreases. This makes sense since more powerful cars tend to be less fuel-efficient.
*   MPG vs Weight: A strong negative relationship between weight and mpg is likely. Heavier cars often have lower fuel efficiency.
*   MPG vs Displacement: A strong negative relationship between displacement and mpg is likely. Displacement refers to the total volume of all the cylinders in an engine. *   MPG vs cylinders: There is a strong negative relationship between cylinders and mpg. Engines with more cylinders have a greater capacity to combust fuel.
